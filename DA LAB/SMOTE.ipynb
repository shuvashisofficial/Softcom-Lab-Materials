```python
import pandas as pd
import numpy as np
import os
import sys
from shutil import copyfile
import os.path
import numpy as np
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array,load_img
from PIL import Image
from sklearn.model_selection import train_test_split
from numpy import load
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pathlib
from tensorflow import keras
import tensorflow as tf
```

```python
imagegen = ImageDataGenerator()
```

```python
# load train data from drive
train_generator=imagegen.flow_from_directory("D://Brain MRI Dataset//Dataset", class_mode="categorical", shuffle=False, batch_size=16, target_size=(256, 256),seed=42)
x=np.concatenate([train_generator.next()[0] for i in range(train_generator.__len__())])
y=np.concatenate([train_generator.next()[1] for i in range(train_generator.__len__())])
print(x.shape)
print(y.shape)
```

```python
#Converting  our color images to a vector
X_train=x.reshape(1802,256*256*3)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sampling_strategy = {i: 200 for i in range(33)}
```

```python
# Assuming y_train contains the target class with [0, 1] where 1 is the minority
sm = SMOTE(sampling_strategy=sampling_strategy)  # This makes the minority class 50% of the majority class, not equal
X_smote, y_smote = sm.fit_resample(X_train, y)
```

```python
X_smote.shape
```

```python
import numpy as np
from keras.preprocessing.image import array_to_img
import os

# Reshape X_smote into images
X_smote_img = X_smote.reshape(-1, 256, 256, 3)  # automatically calculate the number of images

# Assuming you want to save the first image for verification
pil_img = array_to_img(X_smote_img[0] * 255)  # scale the values to [0, 255] range and convert to PIL Image
os.makedirs('D:\\Brain MRI Dataset\\Images', exist_ok=True)  # ensure the directory exists
pil_img.save('D:\\Brain MRI Dataset\\Images\\1.jpg')

print("Image saved successfully at 'D:\\Brain MRI Dataset\\Images\\1.jpg'")
```

```python
#Save all images generated by the SMOTE method to the drive

train_sep_dir='D://Brain MRI Dataset//Augmented//'
```

```python
#Create a "testfolder" if it does not exist on the drive
if not os.path.exists(train_sep_dir):
    os.mkdir(train_sep_dir)
```

```python
#This function return label name
def get_key(val):
    # for key, value in pathos.items():
    #      if val == value:
    #       return key
    for key, value in enumerate(val):
        if value == 1:
            return key
```

```python
for i in range(len(X_smote_img)):
  label=get_key(y_smote[i])
  print(label)
  if not os.path.exists(train_sep_dir + str(label)):
    os.mkdir(train_sep_dir + str(label))
  pil_img = array_to_img(X_smote_img[i]* 255)
  pil_img.save(train_sep_dir + str(label) +'/'+ str(i) + '.jpg')
```
